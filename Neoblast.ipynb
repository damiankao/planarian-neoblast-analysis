## Data clean-up

Lowly expressed transcripts were filtered out by removing any transcript that had less than 10 tags in more than half of the libraries within each dataset (Aboobaker, Rajewsky, Pearson, Graveley). For Rajewsky and Graveley that contained only 3 libraries, this effectively meant for every transcript, at least one library (3 / 2 rounded down is 1) had to have tag count greater or equal to 10. The intersection of each filtered dataset was used in down-stream analysis. 

```python
def filterData(d, thresh, names):
    filterDatasets = {}
    for lib, data in d.items():
        filterDatasets[lib] = []
        for i, row in enumerate(data):
            check = 0
            for col in row:
                if col > thresh:
                    check += 1
                if check > len(row) / 2:
                    filterDatasets[lib].append(tids[i])
                    break
                
    intersection = set(tids[:])
    for lib, data in filterDatasets.items():
        intersection = set(intersection & set(data))
        
    return (filterDatasets, intersection)

filteredLines = {}
filteredLines['intersection'] = []
for i in range(1,51,5):
    filtered, intersection = filterData(datasets, i, tids)
    for lib, data in filtered.items():
        if not filteredLines.has_key(lib):
            filteredLines[lib] = []
            
        filteredLines[lib].append(len(data))
    filteredLines['intersection'].append(len(intersection))

plt.figure(figsize=[10,6])
for lib, data in filteredLines.items():
    plt.plot(range(1,51,5),data)
    
plt.title('# of transcripts remaining at a range of tag count thresholds')
plt.xlabel('tag count threshold')
plt.ylabel('# of transcripts')
plt.legend(filteredLines.keys())

#intersection used for down-stream analysis
filteredTids = list(filterData(datasets,10,tids)[1])
filteredIntersect = np.array([vals[tids.index(tid)] for tid in filteredTids])
```

### Proportion of neoblast/progeny/differentiated in each data set

To calculate the proportions of each cell population:
 - Neoblast proportion in Aboobaker Irradiation dataset = (Intact - Irradiation Day 2) / Intact
 - Neoblast proportion in Aboobaker H2B dataset = (GFP Day 5 - H2B Day 5) / GFP Day 5
 - Neoblast proportion in FACS datasets = X1 / (X1 + X2 + XIns)
 - Progeny proportion in FACS datasets = X2 / (X1 + X2 + XIns)
 - Differentiated proportion in FACS datasets = X3 / (X1 + X2 + XIns)

```python
DS_abIrra_neoProp = []
DS_abH2b_neoProp = []
for i, row in enumerate(DS_aboobaker):
    if row[3] > row[0]:
        DS_abIrra_neoProp.append((i,(row[3] - row[0]) / row[3]))
    else:
        DS_abIrra_neoProp.append((i,-1 * (row[0] - row[3]) / row[0]))
        
    if row[4] > row[7]:
        DS_abH2b_neoProp.append((i, (row[4] - row[7]) / row[4]))
    else:
        DS_abH2b_neoProp.append((i, -1 * (row[7] - row[4]) / row[7]))
    
DS_rj_neoProp = []
DS_rj_progProp = []
DS_rj_diffProp = []
for i, row in enumerate(DS_rajewsky):
    DS_rj_neoProp.append((i, row[0] / sum(row)))
    DS_rj_progProp.append((i, row[1] / sum(row)))
    DS_rj_diffProp.append((i, row[2] / sum(row)))
    
DS_ps_neoProp = []
DS_ps_progProp = []
DS_ps_diffProp = []
for i, row in enumerate(DS_pearson):
    DS_ps_neoProp.append((i, row[2] / sum(row)))
    DS_ps_progProp.append((i, row[0] / sum(row)))
    DS_ps_diffProp.append((i, row[1] / sum(row)))
    
DS_gv_neoProp = []
DS_gv_progProp = []
DS_gv_diffProp = []
for i, row in enumerate(DS_graveley):
    DS_gv_neoProp.append((i, row[2] / sum(row)))
    DS_gv_progProp.append((i, row[0] / sum(row)))
    DS_gv_diffProp.append((i, row[1] / sum(row)))

#neoblast proportions
plt.figure(figsize=[17,7])
for neo in [DS_abIrra_neoProp,DS_abH2b_neoProp,DS_rj_neoProp,DS_ps_neoProp,DS_gv_neoProp]:  
    density = gaussian_kde([x[1] for x in neo])
    xs = np.linspace(-1,1,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))

plt.title('Proportion of neoblast expression in each dataset')
plt.xlabel('Proportion')
plt.ylabel('density')
plt.legend(['Irradiation','H2B','Rajewsky','Pearson','Graveley'])

#progeny proportions
plt.figure(figsize=[17,7])
for neo in [DS_rj_progProp,DS_ps_progProp,DS_gv_progProp]:  
    density = gaussian_kde([x[1] for x in neo])
    xs = np.linspace(0,1,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))

plt.title('Proportion of progeny cell expression in each dataset')
plt.xlabel('Proportion')
plt.ylabel('density')
plt.legend(['Rajewsky','Pearson','Graveley'])

#diff proportions
plt.figure(figsize=[17,7])
for neo in [DS_rj_diffProp,DS_ps_diffProp,DS_gv_diffProp]:  
    density = gaussian_kde([x[1] for x in neo])
    xs = np.linspace(0,1,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))

plt.title('Proportion of differentiated cell expression in each dataset')
plt.xlabel('Proportion')
plt.ylabel('density')
plt.legend(['Rajewsky','Pearson','Graveley'])
```

### Clustering of average neoblast proportions

An attempt to clustering neoblast proportions. This is essentially a 1 dimensional k-means clustering. Gap stiatistic was first calculated for each k value from 1 to 9. The k-value with highest gap statistic was 1. Decided not to go further with this clustering method.

```python
avgProps = []
for i in range(len(DS_rj_neoProp)):
    avgProp = np.mean([DS_rj_neoProp[i][1], DS_ps_neoProp[i][1], DS_abH2b_neoProp[i][1]])
    avgProps.append(avgProp)

plt.figure(figsize=[16,5])
plt.title('Density of average neoblast proportions')
density = gaussian_kde(avgProps)
xs = np.linspace(0,1,200)
density.covariance_factor = lambda : .20
density._compute_covariance()
plt.plot(xs,density(xs))
plt.ylabel('density')
plt.xlabel('average neoblast proportion')
    
avgProps = np.array([avgProps])

dst = sp.spatial.distance.euclidean
def gap_statistics(data, refs=None, nrefs=20, ks=range(1,10)):
    shape = data.shape

    if refs==None:
        tops = data.max(axis=0) # maxima along the first axis (rows)
        bots = data.min(axis=0) # minima along the first axis (rows)
        dists = sp.matrix(sp.diag(tops-bots)) # the bounding box of the input dataset

        # Generate nrefs uniform distributions each in the half-open interval [0.0, 1.0)
        rands = sp.random.random_sample(size=(shape[0],shape[1], nrefs))

        # Adjust each of the uniform distributions to the bounding box of the input dataset
        for i in range(nrefs):
            rands[:,:,i] = rands[:,:,i]*dists+bots
    else:
        rands = refs

    gaps = sp.zeros((len(ks),))   # array for gap statistics (lenth ks)
    errs = sp.zeros((len(ks),))   # array for model standard errors (length ks)
    difs = sp.zeros((len(ks)-1,)) # array for differences between gaps (length ks-1)

    for (i,k) in enumerate(ks): # iterate over the range of k values
        # Cluster the input dataset via k-means clustering using the current value of k
        kmeans = sklearn.cluster.KMeans(n_clusters=k).fit(data)
        (kmc, kml) = kmeans.cluster_centers_, kmeans.labels_

        # Generate within-dispersion measure for the clustering of the input dataset
        disp = sum([dst(data[m,:],kmc[kml[m],:]) for m in range(shape[0])])

        # Generate within-dispersion measures for the clusterings of the reference datasets
        refdisps = sp.zeros((rands.shape[2],))
        for j in range(rands.shape[2]):
            # Cluster the reference dataset via k-means clustering using the current value of k
            kmeans = sklearn.cluster.KMeans(n_clusters=k).fit(rands[:,:,j])
            (kmc, kml) = kmeans.cluster_centers_, kmeans.labels_

            refdisps[j] = sum([dst(rands[m,:,j],kmc[kml[m],:]) for m in range(shape[0])])

        # Compute the (estimated) gap statistic for k
        gaps[i] = sp.mean(sp.log(refdisps) - sp.log(disp))

        # Compute the expected error for k
        errs[i] = sp.sqrt(sum(((sp.log(refdisp)-sp.mean(sp.log(refdisps)))**2) \
                              for refdisp in refdisps)/float(nrefs)) * sp.sqrt(1+1/nrefs)

    # Compute the difference between gap_k and the sum of gap_k+1 minus err_k+1
    difs = sp.array([gaps[k] - (gaps[k+1]-errs[k+1]) for k in range(len(gaps)-1)])

    #print "Gaps: " + str(gaps)
    #print "Errs: " + str(errs)
    #print "Difs: " + str(difs)

    return gaps, errs, difs

def plot_gap_statistics(gaps, errs, difs):
    # Create a figure
    fig = plt.figure(figsize=(16, 4))

    plt.subplots_adjust(wspace=0.35) # adjust the distance between figures

    # Subplot 1
    ax = fig.add_subplot(121)
    ind = range(1,len(gaps)+1) # the x values for the gaps

    # Create an errorbar plot
    rects = ax.errorbar(ind, gaps, yerr=errs, xerr=None, linewidth=1.0)

    # Add figure labels and ticks
    ax.set_title('Clustering Gap Statistics', fontsize=16)
    ax.set_xlabel('Number of clusters k', fontsize=14)
    ax.set_ylabel('Gap Statistic', fontsize=14)
    ax.set_xticks(ind)

    # Add figure bounds
    ax.set_ylim(0, max(gaps+errs)*1.1)
    ax.set_xlim(0, len(gaps)+1.0)

    # Subplot 2
    ax = fig.add_subplot(122)
    ind = range(1,len(difs)+1) # the x values for the difs
    
    max_gap = None
    if len(np.where(difs > 0)[0]) > 0:
        max_gap = np.where(difs > 0)[0][0] + 1 # the k with the first positive dif

    # Create a bar plot
    ax.bar(ind, difs, alpha=0.5, color='g', align='center')

    # Add figure labels and ticks
    if max_gap:
        ax.set_title('Clustering Gap Differences\n(k=%d Estimated as Optimal)' % (max_gap), \
                     fontsize=16)
    else:
        ax.set_title('Clustering Gap Differences\n', fontsize=16)
    ax.set_xlabel('Number of clusters k', fontsize=14)
    ax.set_ylabel('Gap Difference', fontsize=14)
    ax.xaxis.set_ticks(range(1,len(difs)+1))

    # Add figure bounds
    ax.set_ylim(min(difs)*1.2, max(difs)*1.2)
    ax.set_xlim(0, len(difs)+1.0)

    # Show the figure
    plt.show()

#g, e, d = gap_statistics(avgProps.T)
#takes too long to run. Here are the results:
g = [0.68421123, 0.57858073, 0.51908295, 0.45424543, 0.45690356, 0.41330211, 0.41598558, 0.3958362, 0.3928685]
e = [0.00525073, 0.00448198, 0.00665574, 0.00576964, 0.00538981, 0.00493866, 0.00493594, 0.00651397, 0.00719882]
d = [0.11011249, 0.06615352, 0.07060716, 0.00273168, 0.04854012, 0.00225246, 0.02666336, 0.01016651]
plot_gap_statistics(g,e,d)
```

### Grouping average neoblast proportions by threshold

There isn't enough information in the average neoblast proportions for k-means to classify the transcripts. We have to set an educated threshold for extracting a set of potential neoblast transcripts. We looked at the intersection of transcripts in each of the Rajewsky, Pearson, and H2B RNAi datasets that had a neoblast proportion > 0.5. 

We then categorized the transcripts into 2 sources and 3 sources meaning the transcript is > 0.5 in 2 sources or 3 sources.

```python
neoUnion = list(set(rj_neoFilter) | set(ps_neoFilter) | set(h2b_neoFilter))

min2Source = []
min3Source = []
for i in neoUnion:
    count = 0
    if i in rj_neoFilter:
        count += 1
    if i in ps_neoFilter:
        count += 1
    if i in h2b_neoFilter:
        count += 1
        
    if count == 2:
        min2Source.append(i)
        
    if count == 3:
        min3Source.append(i)
        
print 'Number of transcripts with 2 sources:',len(min2Source)
print 'Number of transcripts with 3 sources:' ,len(min3Source)

dataFilter = []
data2Source = []
data3Source = []
for i in range(len(DS_rj_neoProp)):
    avgProp = np.mean([DS_rj_neoProp[i][1], DS_ps_neoProp[i][1], DS_abH2b_neoProp[i][1]])
    if i in min2Source:
        data2Source.append((i,avgProp))
    elif i in min3Source:
        data3Source.append((i,avgProp))
    else:
        dataFilter.append((i,avgProp))
    
data2Source.sort(key = lambda x : x[1], reverse = True)
data3Source.sort(key = lambda x : x[1], reverse = True)
dataFilter.sort(key = lambda x : x[1], reverse = True)

data2Order = [x[0] for x in data2Source]
data3Order = [x[0] for x in data3Source]
dataFilterOrder = [x[0] for x in dataFilter]

d3Struct = []
d3Struct.append(['Rajewsky', [[0, structurify(data3Order, DS_rj_neoProp, DS_rj_progProp, DS_rj_diffProp)], [len(data3Order), structurify(data2Order, DS_rj_neoProp, DS_rj_progProp, DS_rj_diffProp)], [len(data2Order) + len(data3Order), structurify(dataFilterOrder, DS_rj_neoProp, DS_rj_progProp, DS_rj_diffProp)]]])
d3Struct.append(['Pearson', [[0, structurify(data3Order, DS_ps_neoProp, DS_ps_progProp, DS_ps_diffProp)], [len(data3Order), structurify(data2Order, DS_ps_neoProp, DS_ps_progProp, DS_ps_diffProp)], [len(data2Order) + len(data3Order), structurify(dataFilterOrder, DS_ps_neoProp, DS_ps_progProp, DS_ps_diffProp)]]])
d3Struct.append(['Graveley', [[0, structurify(data3Order, DS_gv_neoProp, DS_gv_progProp, DS_gv_diffProp)], [len(data3Order), structurify(data2Order, DS_gv_neoProp, DS_gv_progProp, DS_gv_diffProp)], [len(data2Order) + len(data3Order), structurify(dataFilterOrder, DS_gv_neoProp, DS_gv_progProp, DS_gv_diffProp)]]])
d3Struct.append(['H2B RNAi', [[0, structurifyAB(data3Order, DS_abH2b_neoProp)], [len(data3Order), structurifyAB(data2Order, DS_abH2b_neoProp)], [len(data2Order) + len(data3Order), structurifyAB(dataFilterOrder, DS_abH2b_neoProp)]]])
d3Struct.append(['Irradiation', [[0, structurifyAB(data3Order, DS_abIrra_neoProp)], [len(data3Order), structurifyAB(data2Order, DS_abIrra_neoProp)], [len(data2Order) + len(data3Order), structurifyAB(dataFilterOrder, DS_abIrra_neoProp)]]])
d3Out = open('data2.js','w')
d3Out.write('var data = ' + str(d3Struct) + ';')
d3Out.write('var numTranscripts = ' + str(len(DS_rj_neoProp)) + ';')
d3Out.close()

#don't re-reaster every run
#os.system("phantomjs raster.js index2.html neoProp2.png '1920px*1000px'")
Image(filename='neoProp2.png')
```
