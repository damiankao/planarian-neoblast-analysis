## Neoblast transcripts sequence clustering

Identified neoblast transcripts were taken from the 4 transcriptomes. CAP3 was used to cluster the sequences together. CAP3 meta-data was parsed in to a tab delimited file containing member transcripts.

```python
from collections import Counter
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import seaborn as sns
from scipy.stats import gaussian_kde
%reload_ext rmagic
%matplotlib inline

#plot styles
sns.set_style("darkgrid")

#load fasta file of all input transcripts to get source counts
faFile = open('resource/aboobaker.graveley.pearson.rajewsky.renamed.fa','r')
capIn = []
for line in faFile:
    if line[0] == '>':
        capIn.append(line[1].upper())

neoInCount = Counter(capIn)
print 'Number of neoblast transcripts input into CAP3:'
print '\n'.join([k + ' - ' + str(c) for k, c in neoInCount.items()])
print

#load meta data of the cap3 clustering
capFile = open('resource/cap3.clusters','r')
clusters = []
allMembers = []
for line in capFile:
    data = line.strip().split('\t')
    members = [x[0].upper() for x in data[1].split(',')]
    clusters.append(members)
    allMembers.extend(members)

memberCount = Counter(allMembers)

print 'Number of clusters: ' + str(len(clusters))
print
print 'Number of transcripts from each data source that went into clusters: '
print '\n'.join([k + ' - ' + str(c) for k, c in memberCount.items()])
print
print 'Percentage of neoblast transcripts from each data source that went into clusters: '
print '\n'.join([k + ' - ' + str(float(memberCount[k]) / neoInCount[k] * 100) for k, c in memberCount.items()])

ind = range(4)
plt.figure(figsize=[15,6])
p_neoIn = plt.bar(ind, [v for k,v in neoInCount.items()], 0.5, align='center')
p_clusterCount = plt.bar(ind, [v for k,v in memberCount.items()], 0.5, color='r', align='center')
plt.title('Number of neoblast transcripts from each data source')
plt.xlabel('Dataset')
plt.xticks(range(4), ['Aboobaker','Pearson','Rajewsky','Graveley'], size='large')
plt.ylabel('Number of transcripts')
plt.legend((p_neoIn, p_clusterCount), ('Input neoblast transcripts', 'Contributed to clustering'), loc=2)
```

## Data remapping
Illumina reads were mapped with Bowtie2 to a consolidated transcriptome described in (citation). ABI SOLiD reads were mapped with ABI LifeScope to the consolidated transcriptome. HTseq-count was then used on the mapping files to obtain tag counts of uniquely mapped reads with a mapping quality of more than 30. The raw counts were consolidated into a tab delmited file, `all.count`.

Some general information about this file: 

```python
#load and process data into numpy matrix
data = [row.strip().split('\t') for row in open('resource/all.count','r')]
header = data[0][1:]
tids = [row[0] for row in data[1:]]
vals = np.array([[float(col) for col in row[1:]] for row in data[1:]], dtype=float)

print 'number of transcripts: ' + str(len(tids))
print
print 'number of conditions: ' + str(len(header))
print
print 'conditions: \n' + '\n'.join([str(i + 1) + ' - ' + d for i,d in enumerate(header)])
```

## FACS dataset analysis

Rajewsky, Pearson and Graveley datasets were generated from FACS samples. Each contain RNA-seq on X1, X2, and Xins cell populations. 

The FACS datasets were converted to ranks and a hierarchical clustering was done on both the raw counts and ranks:

```python
#columns for each data source
ab = [7,8,9,10,11,12,13,23,24,25,26]
rj = [4,5,6]
ps = [14,15,16,17,18,19]
gv = [20,21,22]
f = rj + ps + gv
f.sort()

#put data into structures
aboobaker = vals[:,ab]
rajewsky = vals[:,rj]
pearson = vals[:,ps]
graveley = vals[:,gv]
facs = vals[:,f]

datasets = {'aboobaker':aboobaker,'rajewsky':rajewsky,'pearson':pearson,'graveley':graveley}

#normalize FACS dataset in R using edgeR
facsHeader = [header[i] for i in f]
%R library(edgeR)
%R -i facs
%R -i facsHeader
%R DGEData = DGEList(counts=facs,group = facsHeader)
%R DGEData = calcNormFactors(DGEData)
%R normFacs = cpm(DGEData,normalized.lib.sizes=FALSE)
%R -o normFacs

#plot hierarchical clustering of normalized FACS data
dist = pdist(normFacs.T,'correlation')
link = linkage(dist,method='complete')
plt.figure(figsize=[14,6])
dendrogram(link,labels=facsHeader,orientation='left')
plt.xlabel('Distance')
plt.ylabel('Samples')
plt.title('Normalized FACS data clustering', fontweight='bold', fontsize=14);
```

```python
#plot hierarchical clustering of ranked FACS data with filtering of lowly expressed genes
facsFilter = np.array([row for row in facs if sum(row) > 0])
facsRank = np.array([stats.rankdata(con) for con in facsFilter.T])
dist = pdist(facsRank,'correlation')
link = linkage(dist,'complete')
plt.figure(figsize=[14,6])
dendrogram(link,labels=facsHeader,orientation='left')
plt.xlabel('Distance')
plt.ylabel('Samples')
plt.title('Ranked FACS data clustering', fontweight='bold', fontsize=14);
```

## Data clean-up

Lowly expressed transcripts were filtered out by removing any transcript that had less than 10 tags in less than half of the libraries within each dataset (Aboobaker, Rajewsky, Pearson, Graveley). For Rajewsky and Graveley that contained only 3 libraries, this effectively meant for every transcript, at least one library (3 / 2 rounded down is 1) had to have tag count greater or equal to 10. The intersection of each filtered dataset was used in down-stream analysis. 

```python
def filterData(d, thresh, names):
    filterDatasets = {}
    for lib, data in d.items():
        filterDatasets[lib] = []
        for i, row in enumerate(data):
            check = 0
            for col in row:
                if col > thresh:
                    check += 1
                if check > len(row) / 2:
                    filterDatasets[lib].append(tids[i])
                    break
                
    intersection = set(tids[:])
    for lib, data in filterDatasets.items():
        intersection = set(intersection & set(data))
        
    return (filterDatasets, intersection)

filteredLines = {}
filteredLines['intersection'] = []
for i in range(1,51,5):
    filtered, intersection = filterData(datasets, i, tids)
    for lib, data in filtered.items():
        if not filteredLines.has_key(lib):
            filteredLines[lib] = []
            
        filteredLines[lib].append(len(data))
    filteredLines['intersection'].append(len(intersection))

plt.figure(figsize=[10,6])
for lib, data in filteredLines.items():
    plt.plot(range(1,51,5),data)
    
plt.title('# of transcripts remaining at a range of tag count thresholds')
plt.xlabel('tag count threshold')
plt.ylabel('# of transcripts')
plt.legend(filteredLines.keys())

#intersection used for down-stream analysis
filteredTids = list(filterData(datasets,10,tids)[1])
filteredIntersect = np.array([vals[tids.index(tid)] for tid in filteredTids])
```

The tag counts for the filtered intersecting transcripts were normalized using EdgeR.

```python
#normalize using edgeR
edgeRHeader = ['_'.join(lib.split('_')[:-1]) for lib in header]
%R library(edgeR)
%R -i filteredIntersect
%R -i edgeRHeader
%R DGEData = DGEList(counts=filteredIntersect,group = edgeRHeader)
%R DGEData = calcNormFactors(DGEData)
%R normFilteredVals = cpm(DGEData,normalized.lib.sizes=FALSE)
%R -o normFilteredVals

normFilteredRanks = np.array([stats.rankdata(con) for con in normFilteredVals.T])
normFilteredZScore = stats.zscore(normFilteredVals, axis = 1)
print 'Final number of filtered and normalized transcripts: ' + str(len(normFilteredVals))
```

```python
dist = pdist(normFilteredVals.T,'correlation')
link = linkage(dist,method='complete')
plt.figure(figsize=[14,10])
dendrogram(link,labels=header,orientation='left')
plt.xlabel('Distance')
plt.ylabel('Samples')
plt.title('Normalized filtered data clustering', fontweight='bold', fontsize=14);
```

Take the mean of replicates as the working expression value.

```python
normFilterMeanVals = []
for r in normFilteredVals:
    newRow = [np.mean([r[0],r[1]]),np.mean([r[2],r[3]]),r[4],r[5],r[6],np.mean([r[7],r[10]]),np.mean([r[8],r[11]]),np.mean([r[9],r[12]]),r[13],np.mean([r[14],r[15]]),np.mean([r[16],r[19]]),np.mean([r[17],r[18]])]
    newRow.extend(r[20:])
    normFilterMeanVals.append(newRow)

normFilterMeanVals = np.array(normFilterMeanVals)
normFilterMeanHeader = []
for item in edgeRHeader:
    if not item in normFilterMeanHeader:
        normFilterMeanHeader.append(item)
```

### Proportion of neoblast/progeny/differentiated in each data set

To calculate the proportions of each cell population:
 - Neoblast proportion in Aboobaker Irradiation dataset = (Intact - Irradiation Day 2) / Intact
 - Neoblast proportion in Aboobaker H2B dataset = (GFP Day 5 - H2B Day 5) / GFP Day 5
 - Neoblast proportion in FACS datasets = X1 / (X1 + X2 + XIns)
 - Progeny proportion in FACS datasets = X2 / (X1 + X2 + XIns)
 - Differentiated proportion in FACS datasets = X3 / (X1 + X2 + XIns)

```python
DS_abIrra_neoProp = []
DS_abH2b_neoProp = []
for i, row in enumerate(DS_aboobaker):
    if row[3] > row[0]:
        DS_abIrra_neoProp.append((i,(row[3] - row[0]) / row[3]))
    else:
        DS_abIrra_neoProp.append((i,-1 * (row[0] - row[3]) / row[0]))
        
    if row[4] > row[7]:
        DS_abH2b_neoProp.append((i, (row[4] - row[7]) / row[4]))
    else:
        DS_abH2b_neoProp.append((i, -1 * (row[7] - row[4]) / row[7]))
    
DS_rj_neoProp = []
DS_rj_progProp = []
DS_rj_diffProp = []
for i, row in enumerate(DS_rajewsky):
    DS_rj_neoProp.append((i, row[0] / sum(row)))
    DS_rj_progProp.append((i, row[1] / sum(row)))
    DS_rj_diffProp.append((i, row[2] / sum(row)))
    
DS_ps_neoProp = []
DS_ps_progProp = []
DS_ps_diffProp = []
for i, row in enumerate(DS_pearson):
    DS_ps_neoProp.append((i, row[2] / sum(row)))
    DS_ps_progProp.append((i, row[0] / sum(row)))
    DS_ps_diffProp.append((i, row[1] / sum(row)))
    
DS_gv_neoProp = []
DS_gv_progProp = []
DS_gv_diffProp = []
for i, row in enumerate(DS_graveley):
    DS_gv_neoProp.append((i, row[2] / sum(row)))
    DS_gv_progProp.append((i, row[0] / sum(row)))
    DS_gv_diffProp.append((i, row[1] / sum(row)))

#neoblast proportions
plt.figure(figsize=[17,7])
for neo in [DS_abIrra_neoProp,DS_abH2b_neoProp,DS_rj_neoProp,DS_ps_neoProp,DS_gv_neoProp]:  
    density = gaussian_kde([x[1] for x in neo])
    xs = np.linspace(-1,1,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))

plt.title('Proportion of neoblast expression in each dataset')
plt.xlabel('Proportion')
plt.ylabel('density')
plt.legend(['Irradiation','H2B','Rajewsky','Pearson','Graveley'])

#progeny proportions
plt.figure(figsize=[17,7])
for neo in [DS_rj_progProp,DS_ps_progProp,DS_gv_progProp]:  
    density = gaussian_kde([x[1] for x in neo])
    xs = np.linspace(0,1,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))

plt.title('Proportion of progeny cell expression in each dataset')
plt.xlabel('Proportion')
plt.ylabel('density')
plt.legend(['Rajewsky','Pearson','Graveley'])

#diff proportions
plt.figure(figsize=[17,7])
for neo in [DS_rj_diffProp,DS_ps_diffProp,DS_gv_diffProp]:  
    density = gaussian_kde([x[1] for x in neo])
    xs = np.linspace(0,1,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))

plt.title('Proportion of differentiated cell expression in each dataset')
plt.xlabel('Proportion')
plt.ylabel('density')
plt.legend(['Rajewsky','Pearson','Graveley'])
```
