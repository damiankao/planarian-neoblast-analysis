# Mapping reads to transcriptome

```python
import gzip, os, sys
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
from IPython.display import HTML
from collections import defaultdict
```

```python
prefix = '/home/share/projects/smed_neoblast/trimmed/'
fs = set([x.split('.')[0] for x in os.listdir(prefix) if x.split('.')[-1] == 'gz'])
libs = defaultdict(list)
for f in fs:
    libs['_'.join(f.split('_')[:2])].append(f)
    
cmds = ''
peBowtie = 'bowtie2 --phred33 --local -p 3 -x /home/share/transcriptome/smed/ox_smed_v1/bowtie2/ox_Smed_v1.fa -1 LEFT.fq -2 RIGHT.fq'
seBowtie = 'bowtie2 --phred33 --local -p 3 -x /home/share/transcriptome/smed/ox_smed_v1/bowtie2/ox_Smed_v1.fa -U READS.fq'
for lib, f in libs.items():
    if len(f) > 1:
        cmds += peBowtie.replace('LEFT.fq',prefix + f[0] + '.paired.trimmed.fastq.gz').replace('RIGHT.fq',prefix + f[1] + '.paired.trimmed.fastq.gz') \
        + ' > ' + lib + '.sam 2> ' + lib + '.err\n'
    else:
        cmds += seBowtie.replace('READS.fq',prefix + f[0] + '.trimmed.fastq.gz') \
        + ' > ' + lib + '.sam 2> ' + lib + '.err\n'
        
HTML(textarea('Bowtie2 mapping commands',cmds))
```

```python
prefix = '/home/share/projects/smed_neoblast/map/'
bowtie = [x for x in os.listdir(prefix) if x.split('.')[-1] == 'err']
bowtie.sort()

mapTable = ListTable()
mapTable.append(['Libraries','Total reads/fragments','Uniquely Mapped','Multi-Mapped','Uniquely Mapped (%)','Multi-Mapped (%)','Library Type'])
for err in bowtie:
    inFile = open(prefix + err)
    numReads = 0
    unique = 0
    uniqueP = 0
    multi = 0
    multiP = 0
    libType = 'SE'
    for line in inFile:
        if line.find('reads') != -1:
            numReads = int(line.strip().split()[0])
        elif line.find('exactly 1 time') != -1:
            unique = int(line.strip().split()[0])
            uniqueP = line.strip().split()[1][1:-1]
        elif line.find('>1 times') != -1:
            multi = int(line.strip().split()[0])
            multiP = line.strip().split()[1][1:-1]
            
        if line.find('----') != -1:
            libType = 'PE'
            break
    mapTable.append([err.split('.')[0],commas(numReads),commas(unique),commas(multi),uniqueP,multiP,libType])
    inFile.close()
print 'Bowtie2 output'
mapTable
```

## Convert to tag counts

How Bowtie2 determine uniquely mapping is unknown. Checked BioStar and Seqanswers and no one seems to exactly know. I extracted uniquely mapped reads by filtering for mapping quality equal to 44. For PE reads, I looked for sam flags of 99 or 83 which indicates one of the two properly mapped paired reads. 

```python
#convert to tag counts
prefix = '/home/share/projects/smed_neoblast/map/'
outPrefix = '/home/share/projects/smed_neoblast/counts/'
tidHeaders = [x[1:] for x in  open(prefix + 'ox_smed_v1.headers').read().strip().split('\n')]
tidHeaders.sort()

for item in mapTable[1:]:
    run = item[0]
    inFile = open(prefix + run + '.sam')
    outFile = open(outPrefix + run + '.count','w')
    
    countTable = defaultdict(int)
    for line in inFile:
        if line[0] != '@':
            meta = line.strip().split()
            
            #only take map quality = 44. Seems to be highest score bowtie2 gives.
            if int(meta[4]) == 44:
                if item[-1] == 'PE':
                    #if pe, only take 83 or 99 sam flag 
                    if int(meta[1]) == 83 or int(meta[1]) == 99:
                        countTable[meta[2]] += 1
                else:
                    countTable[meta[2]] += 1
    for tid in tidHeaders:
        outFile.write(tid + '\t' + str(countTable[tid]) + '\n')
                
    inFile.close()
    outFile.close()
```

```python
tagFiles = os.listdir('/home/share/projects/smed_neoblast/counts/')

tags = {}
for f in tagFiles:
    name = f.split('.')[0]
    inFile = open('/home/share/projects/smed_neoblast/counts/' + f)
    count = [int(x.split()[1]) for x in inFile.read().strip().split('\n')]
    tags[name] = sum(count)
    inFile.close()
    
tagTable = ListTable()
tagTable.append(['Library','Total reads/fragments','Uniquely Mapped'])
for item in mapTable[1:]:
    tagTable.append([item[0],item[1],commas(tags[item[0]])])
    
tagTable
```

```python
#consolidate count files
prefix = '/home/share/projects/smed_neoblast/counts/'
fs = os.listdir(prefix)

libs = defaultdict(dict)
tids = []
for f in fs:
    name = f.split('.')[0]
    inFile = open(prefix + f)
    for line in inFile:
        meta = line.strip().split()
        libs[name][meta[0]] = meta[1]
        tids.append(meta[0])
headers = libs.keys()
headers.sort()
tids = list(set(tids))
tids.sort()
outFile = open('/home/share/projects/smed_neoblast/all.count','w')
outFile.write('id' + '\t' + '\t'.join(headers) + '\n')
for tid in tids:
    row = [tid]
    for header in headers:
        row.append(libs[header][tid])
    outFile.write('\t'.join(row) + '\n')
outFile.close()
```
