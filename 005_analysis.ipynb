## Hierarchical clustering

Whole library clustering was done with normalized counts vs ranked counts. Only genes with normalized counts greater than 0 for all libraries were used. According to the following paper, ranking performs well among different transformation methods:

    Zwiener, I, Frisch, B, Binder, H (2014). Transforming RNA-Seq data to improve the performance of prognostic gene signatures. PLoS ONE, 9, 1:e85150.

```python
counts = pd.read_csv('/home/share/projects/smed_neoblast/analysis/all.deseq.counts', sep='\t', index_col=0)
neoblasts = counts[[x for x in counts.axes[1] if any(k in x for k in ['aboobaker','graveley','rajewsky','pearson'])]]
```

## Calculate cell population proportions


```python
#neoblast proportions
pearson_neo_prop = df['pearson_x1'] / (df['pearson_x1'] + df['pearson_x2'] + df['pearson_x3'])
graveley_neo_prop = df['graveley_x1'] / (df['graveley_x1'] + df['graveley_x2'] + df['graveley_x3'])
rajewsky_neo_prop = df['rajewsky_x1'] / (df['rajewsky_x1'] + df['rajewsky_x2'] + df['rajewsky_x3'])

def h2b_irra_neo(x):
    diff = x[0] - x[1]
    if diff < 0:
        return diff / x[1]
    else:
        return diff / x[0]
    
h2b_neo_prop = df[['aboobaker_gfp_5d_A','aboobaker_h2b_5d_A']].apply(h2b_irra_neo,axis=1)
irr_neo_prop = df[['aboobaker_control_0d_A','aboobaker_irr_2d']].apply(h2b_irra_neo,axis=1)
```

```python
pearson_neo_prop.plot(kind='kde', figsize=[15,7],xlim=[-1,1])
graveley_neo_prop.plot(kind='kde')
rajewsky_neo_prop.plot(kind='kde')
h2b_neo_prop.plot(kind='kde')
irr_neo_prop.plot(kind='kde')
plt.legend(['Pearson','Graveley','Rajewsky','H2B RNAi','Irradiation'])
```

```python
#progeny/x2 proportions
pearson_prog_prop = df['pearson_x2'] / (df['pearson_x1'] + df['pearson_x2'] + df['pearson_x3'])
graveley_prog_prop = df['graveley_x2'] / (df['graveley_x1'] + df['graveley_x2'] + df['graveley_x3'])
rajewsky_prog_prop = df['rajewsky_x2'] / (df['rajewsky_x1'] + df['rajewsky_x2'] + df['rajewsky_x3'])
```

```python
pearson_prog_prop.plot(kind='kde', figsize=[15,5],xlim=[0,1])
graveley_prog_prop.plot(kind='kde')
rajewsky_prog_prop.plot(kind='kde')
plt.legend(['Pearson','Graveley','Rajewsky'])
```

```python
#differentiated/x3 proportions
pearson_diff_prop = df['pearson_x3'] / (df['pearson_x1'] + df['pearson_x2'] + df['pearson_x3'])
graveley_diff_prop = df['graveley_x3'] / (df['graveley_x1'] + df['graveley_x2'] + df['graveley_x3'])
rajewsky_diff_prop = df['rajewsky_x3'] / (df['rajewsky_x1'] + df['rajewsky_x2'] + df['rajewsky_x3'])
```

```python
pearson_diff_prop.plot(kind='kde', figsize=[15,5],xlim=[0,1])
graveley_diff_prop.plot(kind='kde')
rajewsky_diff_prop.plot(kind='kde')
plt.legend(['Pearson','Graveley','Rajewsky'])
```

```python
proportions = pd.concat([pearson_neo_prop,pearson_prog_prop,pearson_diff_prop, \
                         graveley_neo_prop,graveley_prog_prop,graveley_diff_prop, \
                         rajewsky_neo_prop,rajewsky_prog_prop,rajewsky_diff_prop, \
                         h2b_neo_prop,irr_neo_prop],axis=1)
proportions.columns = ['pearson x1 proportion','pearson x2 proportion','pearson x3 proportion', \
                       'graveley x1 proportion','graveley x2 proportion','graveley x3 proportion', \
                       'rajewsky x1 proportion','rajewsky x2 proportion','rajewsky x3 proportion', \
                       'H2B RNAi neoblast proportion','Irradiation neoblast proportion']
```

```python
d3_rj = proportions[['rajewsky x1 proportion','rajewsky x2 proportion','rajewsky x3 proportion']]
d3_ps = proportions[['pearson x1 proportion','pearson x2 proportion','pearson x3 proportion']]
d3_gv = proportions[['graveley x1 proportion','graveley x2 proportion','graveley x3 proportion']]
d3_irra = proportions['Irradiation neoblast proportion']
d3_h2b = proportions['H2B RNAi neoblast proportion']

def makeStruct(ds, ab = False):
    out = []
    if ab:
        for row in ds:
            r = max(row,0)
            out.append([[0,r],[r,1-r]])
    else:
        for row in ds.iterrows():
            r = row[1]
            out.append([[0,r[0]],[r[0],r[1]],[r[0] + r[1], r[2]]])
    return out

d3Struct = []
d3Struct.append(['Rajewsky', makeStruct(d3_rj)])
d3Struct.append(['Pearson', makeStruct(d3_ps)])
d3Struct.append(['Graveley', makeStruct(d3_gv)])
d3Struct.append(['H2B RNAi', makeStruct(d3_h2b,ab=True)])
d3Struct.append(['Irradiation', makeStruct(d3_irra,ab=True)])
d3Out = open('resource/data.js','w')
d3Out.write('var data = ' + str(d3Struct) + ';')
d3Out.close()
```

### List of neoblast genes

The final list of neoblast genes are generated by:

 1. For each of the Rajewsky, Pearson, Graveley. H2B RNAi neoblast proportion lists, get genes > 0.5 neoblast proportion. This ensures that the majority of the gene expression is in neoblasts.
 
 2. Categorize each gene by membership in the lists defined in the previous step. 3 categories: "In 2 sources", "In 3 sources", "In 4 sources".
 
 3. Genes that are found to be > 0.5 in all 4 sources are highest confidence list. Genes found to be > 0.5 in 3, 2 sources are lower confidence lists. 

```python
neoProp = proportions[['rajewsky x1 proportion','pearson x1 proportion','graveley x1 proportion','H2B RNAi neoblast proportion']]

proportions

neoblastList = {}
for neoThresh in [0.5,0.6,0.7,0.8,0.9]:
    tidCount = defaultdict(list)
    neo_rjy = proportions['rajewsky x1 proportion']
    neo_psn = proportions['pearson x1 proportion']
    neo_gvy = proportions['graveley x1 proportion']
    neo_h2b = proportions['H2B RNAi neoblast proportion']
    
    neo_rjy = set(neo_rjy[neo_rjy >= neoThresh].index)
    neo_psn = set(neo_psn[neo_psn >= neoThresh].index)
    neo_gvy = set(neo_gvy[neo_gvy >= neoThresh].index)
    neo_h2b = set(neo_h2b[neo_h2b >= neoThresh].index)
    union = set(neo_rjy | neo_psn | neo_gvy | neo_h2b)
    
    for tid in union:
        count = 0
        if tid in neo_rjy:
            count += 1
        if tid in neo_psn:
            count += 1
        if tid in neo_gvy:
            count += 1
        if tid in neo_h2b:
            count += 1
            
        tidCount[count].append(tid)
        
    neoblastList[neoThresh] = tidCount
```

```python
neoList_2 = proportions.ix[neoblastList[0.5][2],].ix[propMean.axes[0]].dropna()
neoList_3 = proportions.ix[neoblastList[0.5][3],].ix[propMean.axes[0]].dropna()
neoList_4 = proportions.ix[neoblastList[0.5][4],].ix[propMean.axes[0]].dropna()
neo_used = set(list(neoList_2.axes[0]) + list(neoList_3.axes[0]) + list(neoList_4.axes[0]))
neoList_0 = proportions.ix[[x for x in proportions.axes[0] if x not in neo_used],].ix[propMean.axes[0]].dropna()

d3Struct = []
d3_rj = ['rajewsky x1 proportion','rajewsky x2 proportion','rajewsky x3 proportion']
d3_ps = ['pearson x1 proportion','pearson x2 proportion','pearson x3 proportion']
d3_gv = ['graveley x1 proportion','graveley x2 proportion','graveley x3 proportion']
d3_irra = 'Irradiation neoblast proportion'
d3_h2b = 'H2B RNAi neoblast proportion'

d3Struct.append(['Rajewsky',[[0,makeStruct(neoList_4[d3_rj])],\
                             [len(neoList_4),makeStruct(neoList_3[d3_rj])],\
                             [len(neoList_4) + len(neoList_3),makeStruct(neoList_2[d3_rj])],\
                             [len(neoList_4) + len(neoList_3) + len(neoList_2),makeStruct(neoList_0[d3_rj])]]])
d3Struct.append(['Pearson',[[0,makeStruct(neoList_4[d3_ps])],\
                             [len(neoList_4),makeStruct(neoList_3[d3_ps])],\
                             [len(neoList_4) + len(neoList_3),makeStruct(neoList_2[d3_ps])],\
                             [len(neoList_4) + len(neoList_3) + len(neoList_2),makeStruct(neoList_0[d3_ps])]]])
d3Struct.append(['Graveley',[[0,makeStruct(neoList_4[d3_gv])],\
                             [len(neoList_4),makeStruct(neoList_3[d3_gv])],\
                             [len(neoList_4) + len(neoList_3),makeStruct(neoList_2[d3_gv])],\
                             [len(neoList_4) + len(neoList_3) + len(neoList_2),makeStruct(neoList_0[d3_gv])]]])
d3Struct.append(['H2B RNAi',[[0,makeStruct(neoList_4[d3_h2b],ab=True)],\
                             [len(neoList_4),makeStruct(neoList_3[d3_h2b],ab=True)],\
                             [len(neoList_4) + len(neoList_3),makeStruct(neoList_2[d3_h2b],ab=True)],\
                             [len(neoList_4) + len(neoList_3) + len(neoList_2),makeStruct(neoList_0[d3_h2b],ab=True)]]])
d3Struct.append(['Irradiation',[[0,makeStruct(neoList_4[d3_irra],ab=True)],\
                             [len(neoList_4),makeStruct(neoList_3[d3_irra],ab=True)],\
                             [len(neoList_4) + len(neoList_3),makeStruct(neoList_2[d3_irra],ab=True)],\
                             [len(neoList_4) + len(neoList_3) + len(neoList_2),makeStruct(neoList_0[d3_irra],ab=True)]]])

d3Out = open('resource/data2.js','w')
d3Out.write('var data = ' + str(d3Struct) + ';')
d3Out.write('var numTranscripts = ' + str(len(proportions)) + ';')
d3Out.close()
```
